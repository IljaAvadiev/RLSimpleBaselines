```python
agent = VPG(
    env,
    nS,
    nA,
    (128, 64),
    activation=F.relu,
    optimizer=optim.Adam,
    pi_alpha=0.0005,
    v_alpha=0.0007,
    beta=0.01,
    gamma=0.99,
    dir='tmp', 
    name='CartPole VPG'
           )
```

```python
vpg_rewards_mean = agent.learn(max_episodes=15000, average_len=100, target_reward=200, log=True)
```
