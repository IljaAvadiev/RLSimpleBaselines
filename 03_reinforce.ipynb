# Reinforce (and VPG)

```python
import gym
from algorithms.deeprl.reinforce.agent import Reinforce
import torch.optim as optim
import torch.nn.functional as F
```

```python
env = gym.make('CartPole-v1')
nS = env.observation_space.shape[0]
nA = env.action_space.n
```

```python
agent = Reinforce(
    env, 
    nS, 
    nA, 
    (128, 64), 
    activation=F.relu, 
    optimizer=optim.RMSprop,
    alpha=0.0001, 
    gamma=0.99, 
    dir='tmp', 
    name='CartPole Reinforce')
```

```python
agent.learn(max_episodes=10000, average_len=100, target_reward=475, log=True)
```
